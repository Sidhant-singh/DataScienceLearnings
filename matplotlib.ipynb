```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.random.rand(50)
y= np.random.rand(50)
```

```python
x
```

```python
y
```

```python
plt.figure(figsize=(6,2))
plt.scatter(x,y , c='g',alpha=.9,marker='x')
plt.xlabel("This is my x axis")
plt.ylabel("This is my y axis")
plt.title("X vs Y")
plt.grid()
```

```python
x = np.linspace(1,10,100)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.scatter(x,y)
```

```python
plt.plot(x,y,c='r')
```

```python
plt.figure(figsize=(8,4))
plt.plot(x,y,'--r')
plt.xlabel("X data")
plt.ylabel("Y data")
plt.title("Sin wave graph")
plt.grid()
```

```python
x = ['a','b','c','d','e']
y = np.random.rand(5)
```

```python
x
```

```python
y
```

```python
plt.bar(x,y,color ='b')
plt.xlabel("Categorical data")
plt.ylabel("Numberical data")
plt.grid()
plt.figure(figsize=(5,3))
```

```python
plt.barh(x,y,color ='b') ##barh - bar horizontal
plt.xlabel("Categorical data")
plt.ylabel("Numberical data")
plt.grid()
plt.figure(figsize=(5,3))
```

```python
data = [1,2,3,4,5,6,5,6,6,7,7,8,8,8,8,8]
```

```python
#histogram - will give you the frequency of unique elements.
plt.figure(figsize=(6,3))
plt.hist(data,color='b')
plt.ylabel("Distinct elements")
plt.xlabel("Frequency of elements")
plt.title("This is histogram")
plt.grid()
```

```python
## 3d graph
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection='3d')
ax.scatter(x,y,z)
plt.xlabel("X axis")
plt.ylabel("Y axis")
## plt.zlabel("Z axis")
plt.show()
```
